```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country
```

```python
final_df = pd.merge(df, df_country, on="Country Code", how="left")
final_df
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
## Top 3 Countries
plt.pie(country_val[:3] ,labels=country_names[:3], autopct="%1.2f%%")
plt.show()
```

Observation: Zomato Major transactions are from India, then USA and then UK.

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observation: Wherever the rating is from 4.5 to 4.9 -> Excellent 
1)                                       4.0 to 4.4 -> Very Good
2)                                       3.5 to 3.9 -> good 
3)                                       2.5 to 3.4 -> Average
4)                                       1.8 to 2.4 -> Poor
5)                                       0.0 -> Not Rated

```python
ratings.head()
```

```python
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating", y="Rating Count", hue="Rating color", data=ratings, palette=["#EEEADF", 'red' , 'orange', 'yellow', '#90EE90', '#006400'])
plt.show()
```

Observation: 
1) Not Rated is Very High
2) Max number of Rating is between 2.5 to 3.4

```python
sns.countplot(x="Rating color", data=ratings, palette=["#EEEADF", 'red' , 'orange', 'yellow',  '#90EE90', '#006400'])
plt.show()
```

Count is Frequency of records for ratings. 

## Lets Find out country names that has given zero ratings

```python
# final_df.groupby(['Country','Aggregate rating',"Rating color", "Rating text"]).size()
rating_df = final_df[['Country','Aggregate rating',"Rating color", "Rating text"]]
zero_rating_count_df = rating_df[rating_df['Aggregate rating'] == 0.0 ].groupby('Country')["Aggregate rating"].count().reset_index().copy()
zero_rating_count_df
```

## Find the top 10 cuisines corresponds to their city and country

```python
final_df[['Country', 'City', 'Cuisines']]
cuisines = final_df['Cuisines'].str.split(', ').copy()
cuisines.value_counts()
```

```python
df_exploded = cuisines.explode()
counts = df_exploded.value_counts().head(10)
cuisines_labels = counts.index
cuisines_values = counts.values
```

```python
plt.pie(cuisines_values, labels=cuisines_labels, autopct="%1.2f%%")
plt.show()
```
